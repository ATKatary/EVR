```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
result_dir = './run_2/'

bilin = result_dir + 'val_bilin.png'
ref = result_dir + 'val_ref.png'
res = result_dir + 'val_restored.png'

t = (0, 1, 2)
bilin = plt.imread(bilin).transpose(t)[..., 0:3]
ref = plt.imread(ref).transpose(t)[..., 0:3]
res = plt.imread(res).transpose(t)[..., 0:3]
```

```python
def roi(img):
    return img[0:150, 300:600, :]
```

```python
fig = plt.figure(figsize=(15, 8), dpi=200)

#########################
plt.subplot(131)
#plt.axis('off')
plt.imshow(roi(ref))
plt.xlabel('Ground Truth')

#########################
plt.subplot(132)
plt.imshow(roi(bilin))
#plt.axis('off')
plt.xlabel('Bilinear Baseline')

#######################
plt.subplot(133)
plt.imshow(roi(res))
#plt.axis('off')
plt.xlabel('NSSR Rconstruction')

for ax in fig.axes:
    ax.get_yaxis().set_ticks([])
    ax.get_xaxis().set_ticks([])

plt.tight_layout()
plt.savefig('out.png')
```
